```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
data =sns.load_dataset("iris")
```

```python
data.head()
```

```python
data.species.value_counts()
```

```python
data.isnull().sum()
```

```python
data.info()
```

### BarPlot

```python
sns.barplot(x="species",y="sepal_length",data=data)
```

### CountPlot

```python
sns.countplot(x="species",data=data)
```

### Boxplot

```python
sns.boxplot(data=data)
```

```python
sns.boxplot(x="species",y="sepal_width",data=data)
```

### Violinplot

```python
sns.violinplot(x="species",y="sepal_width",data=data)
```

### Stripplot

```python
sns.stripplot(x="species",y="petal_length",data=data)
```

### Swarmplot

```python
sns.swarmplot(x="species",y="petal_length",data=data)
```

### Distribution Plots

### Distplot

```python
sns.distplot(data.petal_length)
```

```python
 sns.distplot(data.petal_length,kde=False)
```

### Joint Plot:

```python
sns.jointplot(x="sepal_length",y="sepal_width",data=data)
```

### Pairplot:

```python
sns.pairplot(data)
```

 ### Rugplot

```python
sns.rugplot(data.petal_length)
```

 ### Matrix Plot

### Heatmaps:

```python
correlation=data.corr()
correlation
sns.heatmap(correlation,annot=True)
```

### Grids


### PairGrid

```python
 g=sns.PairGrid(data)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(plt.scatter)
```

 ### FacetGrid:

```python
f=sns.FacetGrid(data,col="species")
f.map(plt.hist,"sepal_width")
```

### JointGrid:

```python
j=sns.JointGrid(x="sepal_length",y="sepal_width",data=data)
j.plot(plt.scatter,sns.distplot)
```

 ### Regression Plot

```python
sns.lmplot(x="sepal_length",y="sepal_width",data=data)
```

 ### Relational Plot

```python
 sns.relplot(x="sepal_length",y="sepal_width",data=data,kind="line")
```

### Scatter Plot:

```python
sns.scatterplot(x="sepal_length",y="sepal_width",hue="species",data=data)
```
